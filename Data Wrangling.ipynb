```python
import pandas as pd
banks = pd.read_excel('Banks Data.xlsx')
banks
```

```python
banks_melt = banks.melt(id_vars=['Name'], var_name = 'Years', value_name = 'Amount' )
```

```python
banks_melt
```

```python
banks_melt['Ratios'] = banks_melt['Name'].str.split('-').str.get(1)
```

```python
banks_melt
```

```python
banks_melt['Bank Name'] = banks_melt['Name'].str.split('-').str.get(0)
```

```python
banks_melt
```

```python
banks_melt = banks_melt.drop('Name', axis = 1)
```

```python
banks_melt.head()
```

```python
banks_cleaned = banks_melt.pivot_table(index=['Bank Name','Years'],columns= 'Ratios', values='Amount')
```

```python
banks_cleaned = banks_cleaned.reset_index()
```

```python
banks_cleaned.columns
```

```python
col_drop = [' SUDAN  ', ' SUSP.01/02/16 ', ' SUSP.01/02/16  ', ' SUSP.12/02/18 ',
       ' SUSP.12/02/18  ', ' SUSP.19/03/21 ', ' SUSP.19/03/21  ',
       ' SUSP.28/03/21  ']

banks_cleaned =banks_cleaned.drop(col_drop, axis = 1)
```

```python
banks_cleaned.rename({"NET_CROFIT": "NET_PROFIT"}, inplace=True)
```
